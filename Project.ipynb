```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. ...correlation b/w features....
df.corr()
```

As observed in corr. matrix or heatmap, GLUCOSE followed BMI have the most and 2nd most high correlation with Outcome.


```python
sns.heatmap(df.corr(), annot=True)    
```

```python
for i in df.columns:
  plt.title(i)
  plt.boxplot(df[i],vert=True)
  plt.show()
```
